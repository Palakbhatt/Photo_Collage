```python
tl=cv2.imread('top_left.jpg')
tl=cv2.cvtColor(tl,cv2.COLOR_BGR2RGB)
tr=cv2.imread('top_right.jpg')
tr=cv2.cvtColor(tr,cv2.COLOR_BGR2RGB)
bl=cv2.imread('bottom_left.jpg')
bl=cv2.cvtColor(bl,cv2.COLOR_BGR2RGB)
br=cv2.imread('bottom_right.jpg')
br=cv2.cvtColor(br,cv2.COLOR_BGR2RGB)
c=cv2.imread('center.jpeg')
c=cv2.cvtColor(c,cv2.COLOR_BGR2RGB)
```

```python
for i in [tl,tr,c,bl,br]:
    plt.imshow(i)
    plt.show()
```

```python
tr=cv2.resize(tr,(200,200))
tl=cv2.resize(tl,(200,200))
br=cv2.resize(br,(200,200))
bl=cv2.resize(bl,(200,200))
ck=cv2.resize(c,(100,100))
```

```python
tl=cv2.copyMakeBorder(tl,10,5,10,5,cv2.BORDER_CONSTANT)
tr=cv2.copyMakeBorder(tr,10,5,5,10,cv2.BORDER_CONSTANT)
ck=cv2.copyMakeBorder(ck,10,10,10,10,cv2.BORDER_CONSTANT)
bl=cv2.copyMakeBorder(bl,5,10,10,5,cv2.BORDER_CONSTANT)
br=cv2.copyMakeBorder(br,5,10,5,10,cv2.BORDER_CONSTANT)
```

```python
for i in [tl,tr,ck,bl,br]:
    plt.imshow(i)
    plt.show()
```

```python
t=np.hstack((tl,tr))
b=np.hstack((bl,br))
p=np.vstack((t,b))
plt.imshow(p)
```
